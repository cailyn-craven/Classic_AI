## Binary Tree 
Create a class for a node in a binary tree of integers. The node should have the following properties and methods.

class Node:

     value

     leftChild

     rightChild

     parent

     getChildren() #returns a list of both children

```python
class Node:
  def __init__(self, value, left, right, parent):
    self.value = value 
    self.leftChild = left
    self.rightChild = right
    self.parent = parent 
  # returns a list of both children 
  def getChildren(self):
    children = []
    children.append(self.leftChild)
    children.append(self.rightChild)
    return children 
```
